# Solution-1 --> Without Considering RBD approach

```python
import pandas as pd
import numpy as np
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
path = '/home/ashish/projects/RBD_AirTrafficControllerSystem/RBD.xlsx'
data_frame = pd.read_excel(path)
data_frame
```

```python
id_vars = ['index']
value_vars = ['System A','System B','System C']
melted_data = pd.melt(data_frame.reset_index(),id_vars,value_vars)
melted_data.columns = ['index','treatments','values']
melted_data
```

```python
model = ols('values ~ C(treatments)',data=melted_data).fit()
model.summary()
```

```python
anova_table = sm.stats.anova_lm(model,typ=1)   ## One way linear anova model
anova_table
```

SAY, Alpha = 5% = 0.05
Since, 

        (p-value = 0.06) > (ALPHA = 0.05)
## So this implies ACCEPT THE NULL HYPOTHESIS/ DO NOT REJECT THE NULL HYPOTHESIS(ALL MEANS ARE EQUAL).

# METHOD::2 -->Considering the RBD APPROACH that is removing the extraneous factor OR nuisance factor present in Error sum of Square previously.

```python
import pandas as pd
import numpy as np
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
path = '/home/ashish/projects/RBD_AirTrafficControllerSystem/RBD.xlsx'
data_frame2 = pd.read_excel(path)
data_frame2
```

```python
id_vars = ['index']
value_vars = ['System A','System B','System C']
melted_data2 = pd.melt(data_frame2.reset_index(),id_vars,value_vars)
melted_data2.columns = ['blocks','treatments','values']
melted_data2
```

```python
model = ols('values ~ C(blocks) + C(treatments)',data = melted_data2).fit()
model.summary()
```

```python
anova_table = sm.stats.anova_lm(model,typ=1)
anova_table
```

## SAY, Alpha = 5% = 0.05 Since,

    (p-value = 0.0241) < (ALPHA = 0.0573)
## So this implies REJECT THE NULL HYPOTHESIS (ALL MEANS ARE EQUAL).

# So this is our most efficient and more dependable result as it has removed the extraneous/nuisance factors while calculating the Error Sum of Square.
